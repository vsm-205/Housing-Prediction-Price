```python
# Data Cleaning

import pandas as pd

raw_data = pd.read_csv(r'D:\Real-Estate-AI-Assitant\data\raw_dataset.csv')

# EDA
# print(raw_data.head(5))
# print(raw_data.info())
# print(raw_data.describe())

# Checking Missing Values
# print(raw_data.isna().sum())

raw_data = raw_data.drop(['Agen','Kantor Agen','No','Judul'], axis=1)
raw_data[['Lokasi','Kota']] = raw_data['Lokasi'].str.split(',', expand=True)

# Apply .str.upper() to each column individually
for col in ['Harga','Lokasi','Kota']:
    raw_data[col] = raw_data[col].str.upper()


# Checking Missing Values
# print(raw_data.isna().sum())
# print(raw_data[raw_data['Garasi'].isna()].head(5))

# Replace Missing Values

raw_data['Garasi'] = raw_data['Garasi'].fillna(value=0, axis=0)
```

```python
# Cleaning the Price Column

def string_to_int_price(rupiah_string):
  cleaning = rupiah_string.replace('RP','').strip() # Replace Rp with space, then remove all the spaces with strip()

  multipliers = {
      'MILIAR' : 1_000_000_000,
      'MILYAR' : 1_000_000_000,
      'JUTA' : 1_000_000,
      'JT' : 1_000_000
  }

  cleaning = cleaning.replace(',','.') # Replace , for . as decimal point

  for word,multiplier in multipliers.items():
    if word in cleaning:
      number = float(cleaning.replace(word,'').strip()) # Take out the multipliers to get the integer values and change it into Float
      return int(number * multiplier)
```

```python
raw_data['Harga'] = raw_data['Harga'].apply(string_to_int_price) # Apply the funtion into the entire dataset column
```

```python
# Cleaning LuasTanah LuasBangunan

def string_to_int_area(area_string):
  cleaning = area_string.replace('m²','').strip() # Replace m² with space, then remove all the spaces with strip()
  return int(cleaning)
```

```python
raw_data['Luas Tanah'] = raw_data['Luas Tanah'].apply(string_to_int_area)
raw_data['Luas Bangunan'] = raw_data['Luas Bangunan'].apply(string_to_int_area)
```

```python
raw_data['Garasi'] = raw_data['Garasi'].astype(int) # Changing Garasi into an Int rather than Decimal
raw_data['Kota'] = raw_data['Kota'].astype(pd.StringDtype()) # Changing Kota into a String
raw_data['Lokasi'] = raw_data['Lokasi'].astype(pd.StringDtype()) # Changing Lokasi into a String

```
